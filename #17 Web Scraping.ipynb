```python
from bs4 import BeautifulSoup
import requests
```

```python
#load a page
r = requests.get("https://www.yellowpages.com/search?search_terms=coffee&geo_location_terms=Washington%2C+DC")
```

```python
#display the content
r.content
```

```python
#declare a soup variable and read in the content
soup = BeautifulSoup(r.content, "lxml")
```

```python
soup
```

```python
#prettify the content
print(soup.prettify())
```

```python
#list all the anchor "a" tags
soup.find_all("a")
```

```python
#show all the links
for link in soup.find_all("a"):
    print(link)
```

```python
#show all the href attributes
for link in soup.find_all("a"):
    print(link.get("href"))
```

```python
#show the text and the link
for link in soup.find_all("a"):
    print(link.text, link.get("href"))
```

```python
#get all the links in proper format
for link in soup.find_all("a"):
    print("<a href='%s'>%s</a>" %(link.get("href"), link.text))
```

```python
#get all the "info" divs
g_data = soup.find_all("div", attrs={"class":"info"})
```

```python
print(g_data)
```

```python
#cleanup the result a bit
for item in g_data:
    print(item.text)
```

```python
#get the items as a list
for item in g_data:
    print(item.contents)
```

```python
#get the first 2 items
items_list = []
for item in g_data:
    items_list.append(item)
items_list[0:2]
```

```python
#print the 0th element's content and text
for item in g_data:
    print(item.contents[0])

for item in g_data:
    print(item.contents[0].text)
```

```python
#get the name and address of the places
for item in g_data:
    print(item.contents[0].text)
    print(item.contents[1].text)
```

```python
#get more specific using the class and build a list
for item in g_data:
    print(item.contents[0].find_all("a", attrs={"class":"business-name"}))
```

```python
#make a pretty list of location names and addresses
for item in g_data:
    print(item.contents[0].find_all("a", attrs={"class":"business-name"})[0].text)
    print(item.contents[1].find_all("p", attrs={"class":"adr"})[0].text)
    print(item.contents[1].find_all("div", {"itemprop":"telephone"})[0].text)
    print()
```
