```python
phish_set_indexed = phish_set.set_index(['target'])
phish_set_indexed.head()
```

```python
phish_set_streadmlined = phish_set_indexed.drop(["phish_id","url",
                                                 "phish_detail_url",
                                                 "submission_time",
                                                 "verification_time"], axis=1)
phish_set_streadmlined
```

```python
phish_set_final = phish_set.drop(["phish_id","url",
                                  "phish_detail_url",
                                  "submission_time",
                                  "verification_time",
                                  "online"], axis=1)
```

```python
phish_set_finalized = phish_set_final.groupby(["target"]).count()
```

```python
phish_set_finalized
```

```python
phish_set_use = phish_set_finalized[phish_set_finalized["verified"] >= 20].drop("Other")
phish_set_use.head()
```

```python
phish_set_use = phish_set_use.rename(columns={"verified":"breaches"})
phish_set_use.head()
```

```python
phish_set_use.count()
```

```python
phish_set_use.plot(kind="bar", figsize=(16,10), legend=True, color="navy", fontsize=14)
```

```python
colorset = ['#0074D9','#7FDBFF','#39CCCC','#3D9970','#85144B','#2ECC40',
            '#FFDC00','#FF851B','#FF4136','#00FF80','#B10DC9','#01FF70']
phish_set_use.plot.pie(subplots=True, figsize=(16,16), legend=False,
                       autopct='%0.2f%%', colors=colorset, fontsize=12)
```

```python
phish_set_by_year = phish_set.drop(['url','phish_detail_url',
                                   'verified','verification_time',
                                   'online'], axis=1)
phish_set_by_year.head()
```

```python
phish_set_by_year = phish_set_by_year.set_index(['target'])
```

```python
phish_set_by_year['submission_time'].dtype
```

```python
phish_set_by_year['submission_time'].head()
```

```python
phish_set_by_year['year'] = phish_set_by_year['submission_time'].astype('datetime64[ns]').dt.year
```

```python
phish_set_by_year.head()
```

```python
phish_set_by_year['year'].dtype
```

```python
phish_set_by_year_group = phish_set_by_year.groupby(['year']).count().drop(
    ['submission_time'], axis=1).rename(columns={'phish_id':'annual_breaches'})
```

```python
phish_set_by_year_group
```

```python
phish_set_by_year_group.plot(kind='barh', legend=True, figsize=(16,10))
```

```python
phish_set_by_year_group.plot(kind='line', legend=True, figsize=(16,10))
```

```python
phish_set_by_month = phish_set.drop(['url','phish_detail_url',
                                   'verified','verification_time',
                                   'online'], axis=1)
phish_set_by_month.head()
```

```python
phish_set_by_month = phish_set_by_month.set_index(['target'])
```
